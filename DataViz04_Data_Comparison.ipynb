### เปรียบเทียบข้อมูล ด้วย coordinate plot

```python
import pandas as pd
```

```python
example_df = pd.read_csv('https://raw.github.com/pandas-dev/''pandas/master/pandas/tests/io/data/csv/iris.csv')
example_df
```

```python
sample_data = example_df.groupby('Name').mean() # หาค่า mean ในเเต่ละ Name
```

```python
sample_data # สุ่ม data
```

```python
sample_data.iloc[0,:]
```

```python
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3,4],sample_data.iloc[0,:],'ro') # ให้กราฟ เป็นสีแดง มีจุดเป็น วงกลม
plt.plot([1,2,3,4],sample_data.iloc[1,:],'g*') # ให้กราฟ เป็นสีเขียว มีจุดเป็น ดาว
plt.plot([1,2,3,4],sample_data.iloc[2,:],'b^') # ให้กราฟ เป็นสีน้ำเงิน มีจุดเป็น สามเหลี่ยม
```

```python
plt.plot([1,2,3,4],sample_data.iloc[0,:],'ro-')
plt.plot([1,2,3,4],sample_data.iloc[1,:],'g*-')
plt.plot([1,2,3,4],sample_data.iloc[2,:],'b^-')
```

```python
plt.plot([1,2,3,4],sample_data.iloc[0,:],'ro:')
plt.plot([1,2,3,4],sample_data.iloc[1,:],'g*:')
plt.plot([1,2,3,4],sample_data.iloc[2,:],'b^:')
```

```python
plt.plot([1,1],[min(sample_data.iloc[:,0]),max(sample_data.iloc[:,0])],'c:')
plt.plot([2,2],[min(sample_data.iloc[:,1]),max(sample_data.iloc[:,1])],'c:')
plt.plot([3,3],[min(sample_data.iloc[:,2]),max(sample_data.iloc[:,2])],'c:')
plt.plot([4,4],[min(sample_data.iloc[:,3]),max(sample_data.iloc[:,3])],'c:') # ลากดูจุด min max ของสีเเดง o

plt.plot([1,2,3,4],sample_data.iloc[0,:],'ro')# ให้กราฟ เป็นสีแดง มีจุดเป็น วงกลม 
plt.plot([1,2,3,4],sample_data.iloc[1,:],'g*') # ให้กราฟ เป็นสีเขียว มีจุดเป็น ดาว  
plt.plot([1,2,3,4],sample_data.iloc[2,:],'b^') # ให้กราฟ เป็นสีน้ำเงิน มีจุดเป็น สามเหลี่ยม
```

## การใช้ plt.plot

Marker -> 'cML'

```python
plt.plot([1],[2],'ro')  # ค่าแกน X เป็น 1 Y เป็น 2  marker r = สีเเดง o คือ วงกลม
```

```python
plt.plot([1,2],[2,4],'ro-.') # กำหนดพร็อต สองจุด
```

```python
plt.plot([1,1],[min(sample_data.iloc[:,0]),max(sample_data.iloc[:,0])],'c:')
plt.plot([2,2],[min(sample_data.iloc[:,1]),max(sample_data.iloc[:,1])],'c:')
plt.plot([3,3],[min(sample_data.iloc[:,2]),max(sample_data.iloc[:,2])],'c:')
plt.plot([4,4],[min(sample_data.iloc[:,3]),max(sample_data.iloc[:,3])],'c:')

plt.plot([1,2,3,4],sample_data.iloc[0,:],'ro',label='Setosa') # ให้กราฟ เป็นสีแดง มีจุดเป็น วงกลม  label ชื่อว่า Setosa
plt.plot([1,2,3,4],sample_data.iloc[1,:],'g*',label='Versicolor') # ให้กราฟ เป็นสีเขียว มีจุดเป็น ดาว  label ชื่อว่า Versicolor
plt.plot([1,2,3,4],sample_data.iloc[2,:],'b^',label='Virginica')  # ให้กราฟ เป็นสีน้ำเงิน มีจุดเป็น สามเหลี่ยม  label ชื่อว่า Virginica
plt.legend()
```

### เปรียบเทียบปริมาณ มากกว่า 1 ปริมาณของ object หลาย class

```python
import numpy as np
```

```python
width = 0.2 #ความกว้างของแท่งกราฟ ต้องกำหนด

plt.bar([1,2],[sample_data['PetalLength'][1],sample_data['PetalWidth'][1]],width,color='g',label ='versicolor') #ดอกที่ 2 versicolor
plt.bar([1-width,2-width],[sample_data['PetalLength'][0],sample_data['PetalWidth'][0]],width,color='r',label ='setosa')
plt.bar([1+width,2+width],[sample_data['PetalLength'][2],sample_data['PetalWidth'][2]],width,color='b',label ='virginica')
plt.xticks([1,2], ['Petal Length','Petal Width'])
plt.legend();
```

```python
plt.bar([1,2,3],sample_data['PetalLength'],color='b',label='Petal Length') #ชุดเเรก
plt.bar([1,2,3],sample_data['PetalWidth'],bottom=sample_data['PetalLength'],color='r',label='Petal Width')
plt.xticks([1,2,3], sample_data.reset_index()['Name'])
plt.grid(True,axis='y') #ใส่เส้นเฉพาะเเกน y
plt.legend();
```

### Bar Chart แบบตะแคงข้าง

```python
plt.barh([1,2,3],sample_data['PetalLength'],color='b',label='Petal Length')
plt.barh([1,2,3],sample_data['PetalWidth'],left=sample_data['PetalLength'],color='r',label='Petal Width')  #เปลี่ยนจาก bottom เป็น left 
plt.yticks([1,2,3], sample_data.reset_index()['Name'])
plt.grid(True,axis='x')
plt.legend();
```

1. คำนวณ อายุเฉลี่ย ของผู้ชาย ของข้อมูลทั้งหมด

```python
data1 = data_covid[['sex','age','nationality','province_of_onset']]
data1
```

```python
male = data1[data1['sex']=='ชาย']
male['age']
```

```python
sum = 0 #ให้เป็น 0 เพราะเริ่มต้นถ้าไม่มีอะไรบวกเข้ามามันต้องเป็น 0 ก่อน 
N = 0 # N นับว่าเป็นกี่คน 
for a in male['age']:
    
    if a > 0: #ถ้ามันมากกว่า 0ก็จะรวมเข้าไปใน sum
        sum += a # sum = sum + a เอา a มาบวกเข้าไปใน sum เรื่อยๆ
        N += 1 #ก็คือบวก 1 เข้าไปเรื่อยๆ ก็คือมี 1 2 3 4 5 6 ตัวเรื่อยๆ

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย {sum/N}')
```

2. คำนวณ อายุเฉลี่ย ของผู้ชาย ของผู้ป่วยในจังหวัดขอนแก่น

```python
male_KK = male[male['province_of_onset']=='ขอนแก่น']
```

```python
sum = 0 #ให้เป็น 0 เพราะเริ่มต้นถ้าไม่มีอะไรบวกเข้ามามันต้องเป็น 0 ก่อน 
N = 0 # N นับว่าเป็นกี่คน 
for a in male_KK['age']: 
    
    if a > 0: #ถ้ามันมากกว่า 0ก็จะรวมเข้าไปใน sum 
        sum += a # sum = sum + a เอา a มาบวกเข้าไปใน sum เรื่อยๆ
        N += 1 #ก็คือบวก 1 เข้าไปเรื่อยๆ ก็คือมี 1 2 3 4 5 6 ตัวเรื่อยๆ

print(f'อายุเฉลี่ย ของ ผู้ป่วยชาย ในจังหวัดขอนแก่น {sum/N}')
```

## HW กลุ่ม 2 || เลือกข้อมูลมาเฉพาะวัน announce_date เป็นวันที่ 1-20

```python
TF2=list()  #สร้าง list 
for each_row in data_covid.iterrows(): #วนลูป
    if each_row[1]['announce_date'].split('/')[1] <= '20':  # split  โดยจะกำหนดเครื่องหมาย /  เป็นตัวคั่นในการ split คือ split หลัง / ตัวที่ 1
        TF2.append(True)  #สร้าง List  ก่อนที่จะเติมค่าลงไปใน List ด้วยการใช้ .append()
    else:
        TF2.append(False)

data_covid[TF2]  #แสดงค่าทั้งหมด
```

###แก้ไข user, site, ov และPartitionKey เป็นตัวเลข 5 หลัก

```python
kk1=list() ## กำหนดให้เป็นลิสต์ในการเติมค่าloop
for each_row in alldb3.iterrows(): ## ทำการวนลูปในแถว
    if len(str(each_row[1]['user']))== 5 : #สร้างเงื่อนไขให้ user มีจำนวนตัวอักษร 5 ตัว
        if str(each_row[1]['user']) <= '99999' : #ถ้า user มีค่า น้อยกว่า 100000
          kk1.append(True) ## ถ้าตรงตามเงื่อนไขให้เพิ่มเข้าไป
        elif str(each_row[1]['user']) >= '99999' : #ถ้า user มีค่า มากกว่า 100000
            kk1.append(False) ## ถ้าไม่ตรงตามเงื่อนไขไม่ให้เพิ่มเข้าไป
    else:
        kk1.append(False) ## ถ้่าไม่ตรงตามเงื่อนไขทั้งหมด ไม่ให้เพิ่มเข้าไป

alldb3[kk1] ## แสดงผล
```

```python
data1 = alldb3[kk1]
```

```python
set(data1['user'])
```

```python
kk2=list() ## กำหนดให้เป็นลิสต์ในการเติมค่าloop
for each_row in data1.iterrows(): ## ทำการวนลูปในแถว
    if len(str(each_row[1]['site']))== 5 : #สร้างเงื่อนไขให้ site มีจำนวนตัวอักษร 5 ตัว
        if str(each_row[1]['site']) <= '99999' : #ถ้า site มีค่า น้อยกว่า 100000
          kk2.append(True) ## ถ้าตรงตามเงื่อนไขให้เพิ่มเข้าไป
        elif str(each_row[1]['site']) >= '99999' : #ถ้า site มีค่า มากกว่า 100000
            kk2.append(False) ## ถ้าไม่ตรงตามเงื่อนไขไม่ให้เพิ่มเข้าไป
    else:
        kk2.append(False) ## ถ้่าไม่ตรงตามเงื่อนไขทั้งหมด ไม่ให้เพิ่มเข้าไป

data1[kk2] ## แสดงผล
```

```python
data2 = data1[kk2]
```

```python
kk3=list() ## กำหนดให้MSเป็นลิสต์ในการเติมค่าloop
for each_row in data2.iterrows(): ## ทำการวนลูปในแถวของ data8
    if len(str(each_row[1]['PartitionKey']))== 5 : #สร้างเงื่อนไขให้ PartitionKey มีจำนวนตัวอักษร 5 ตัว
        if str(each_row[1]['PartitionKey']) <= '99999' : #ถ้า PartitionKey มีค่า น้อยกว่า 100000
          kk3.append(True) ## ถ้าตรงตามเงื่อนไขให้เพิ่มเข้าไปใน MS
        elif str(each_row[1]['PartitionKey']) >= '99999' : #ถ้า PartitionKey มีค่า มากกว่า 100000
            kk3.append(False) ## ถ้าไม่ตรงตามเงื่อนไขไม่ให้เพิ่มเข้าไปใน MS
    else:
        kk3.append(False) ## ถ้่าไม่ตรงตามเงื่อนไขทั้งหมด ไม่ให้เพิ่มเข้าไปใน MS

data2[kk3] ## แสดงผล
```
